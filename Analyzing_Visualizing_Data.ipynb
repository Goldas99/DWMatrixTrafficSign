```python
df_train = pd.read_pickle('data/train.p')
df_test = pd.read_pickle('data/test.p')
```

```python
X_train, y_train = df_train["features"], df_train["labels"]
X_test, y_test = df_test["features"], df_test["labels"]
X_train.shape, y_train.shape
```

```python
signs_names = pd.read_csv('data/signnames.csv')
dict_sign_names = signs_names.to_dict()['b']
```

```python
plt.imshow(X_train[20000])
```

```python
for sign_id in dict_sign_names.keys():
  chosen_sign = X_train[ y_train == sign_id ]

  plt.figure(figsize = (16,6))
  for i in range(5):
      plt.subplot('19{}'.format(i+1))
      plt.imshow( chosen_sign[i] )
      plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
cnt = Counter(y_train).most_common()
id_labels, cnt_labels = zip(*cnt)
ids = range(len(id_labels))
labels = [dict_sign_names[id_labels[id]] for id in id_labels]

plt.figure(figsize = (16,6))
plt.title('Number of data for each class')
plt.bar(ids, cnt_labels)
plt.xlabel('Signs')
plt.xticks(ids, labels, rotation = 'vertical')
plt.ylabel('Amount of data')
```
